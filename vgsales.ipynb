```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

Data cleaning

```python
df=pd.read_csv("vgsales.csv")
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

EDA

```python
plt.figure(figsize=(16,5))
ax=sns.countplot(data=df,x='Year')
for i in ax.containers:
    ax.bar_label(i)
plt.xticks(rotation=90)
plt.show()
```

The most games were produced in the year 2009

```python
sns.countplot(data=df,x='Platform')
plt.xticks(rotation=90)
plt.show()
```

Most games were produced on DS and PS2 platform

```python
sns.countplot(data=df,x='Genre')
plt.xticks(rotation=90)
plt.show()
```

The most games were of Action and sports genre

```python
dfpg=sns.PairGrid(data=df,x_vars='Global_Sales',y_vars=['NA_Sales','EU_Sales','JP_Sales','Other_Sales'])
dfpg.map(plt.scatter)
```

```python
a=df.Publisher.value_counts().head(10).reset_index()
```

```python
sns.barplot(data=a,x='Publisher',y='count')
plt.xticks(rotation=90)
plt.show()
```

the most games were publsihed by Electronic arts and Activision

```python
nasales=df.groupby(df['Publisher'],as_index=False)['NA_Sales'].sum().sort_values(by='NA_Sales',ascending=False).head(10)
sns.barplot(data=nasales,x='Publisher',y='NA_Sales')
plt.xticks(rotation=90)
plt.show()
```

The highest sales in North America region were by Nintendo followed by Electronic arts

```python
eusales=df.groupby(df['Publisher'],as_index=False)['EU_Sales'].sum().sort_values(by='EU_Sales',ascending=False).head(10)
sns.barplot(data=eusales,x='Publisher',y='EU_Sales')
plt.xticks(rotation=90)
plt.show()
```

The highest sales in European region were by Nintendo and Electronic arts


```python
jpsales=df.groupby(df['Publisher'],as_index=False)['JP_Sales'].sum().sort_values(by='JP_Sales',ascending=False).head(10)
sns.barplot(data=jpsales,x='Publisher',y='JP_Sales')
plt.xticks(rotation=90)
plt.title("Japan sales")
plt.show()
```

The highest sales were recorded by Nintendo in Japan region

```python
othersales=df.groupby(df['Publisher'],as_index=False)['Other_Sales'].sum().sort_values(by='Other_Sales',ascending=False).head(10)
sns.barplot(data=othersales,x='Publisher',y='Other_Sales')
plt.xticks(rotation=90)
plt.show()
```

The highest amount of sales in other regions were recorded by Electronic arts

```python
globalsales=df.groupby(df['Publisher'],as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales',ascending=False).head(10)
sns.barplot(data=globalsales,x='Publisher',y='Global_Sales')
plt.xticks(rotation=90)
plt.show()
```

The highest global sales were recorded by Nintendo followed by EA and Activision

```python
genresales=df.groupby(df['Genre'],as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales',ascending=False).head(10)
sns.barplot(data=genresales,x='Genre',y='Global_Sales')
plt.xticks(rotation=90)
plt.show()
```

The most sales was made by Action genre globally

### Conclusion

From all the above analysis we can declare that the highest sellers of Videogames were Nintendo, Electronic Arts and Activision. With Nintendo recording the highest amount of sales Globally. And the Largest producer of Video games among all is Electronic Arts.
