```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Data cleaning

```python
df=pd.read_csv("vgsales.csv")
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

EDA

```python
plt.figure(figsize=(16,5))
ax=sns.countplot(data=df,x='Year')
for i in ax.containers:
    ax.bar_label(i)
plt.xticks(rotation=90)
plt.show()
```

The most games were produced in the year 2009

```python
sns.countplot(data=df,x='Platform')
plt.xticks(rotation=90)
plt.show()
```

Most games were produced on DS and PS2 platform

```python
sns.countplot(data=df,x='Genre')
plt.xticks(rotation=90)
plt.show()
```

The most games were of Action and sports genre

```python
sns.lmplot(data=df,x='NA_Sales',y='EU_Sales')
```
